```python
import pybullet as pb

from evo_motion.model import ObjShape
```

```python
pb.connect(pb.DIRECT)
```

```python
obj = ObjShape("/home/samuel/PycharmProjects/EvoMotion/resources/obj/cube.obj")
```

```python
vertices, indices = obj.get_vertices_indices()
col_shape_1 = pb.createCollisionShape(shapeType=pb.GEOM_MESH, vertices=vertices, indices=indices)
col_shape_2 = pb.createCollisionShape(shapeType=pb.GEOM_MESH, vertices=vertices, indices=indices)
```

```python
col_shape_1, col_shape_2
```

```python
body_1 = pb.createMultiBody(baseCollisionShapeIndex=col_shape_1, basePosition=[0, 0, -1], baseMass=0.)
body_2 = pb.createMultiBody(baseCollisionShapeIndex=col_shape_2, basePosition=[0., 0, 1], baseMass=1.)
```

```python
pb.setGravity(0, 0, -10)
```

```python
for i in range(10000):
    pb.stepSimulation()
    print(pb.getBasePositionAndOrientation(body_1), pb.getBasePositionAndOrientation(body_2))
```
